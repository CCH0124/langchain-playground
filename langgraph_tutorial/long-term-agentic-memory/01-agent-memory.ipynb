## What are types of memory ?

| 記憶類型 (Memory Type)  | 儲存內容 (What is Stored) | 人類例子 (Human Example)                   |
| ------------------- | --------------------- | -------------------------------------- |
| **Semantic 語意記憶**   | 事實 (Facts)            | 我在學校學到的知識 (Things I learned in school) |
| **Episodic 情節記憶**   | 經驗 (Experiences)      | 我做過的事 (Things I did)                   |
| **Procedural 程序記憶** | 指令、步驟 (Instructions)  | 直覺反應或動作技能 (Instincts or motor skills)  |

## Memories can be saved using different mechanisms



In the hot path: 每次收到用戶查詢時都會讀取和寫入記憶體。

In the background: 在代理運行期間，記憶體會被定期讀取和寫入。

下圖為一個使用 Memory 的 Agent Workflow



Episodic Memory: 放在 Prompt 中的 Few-Shot Examples，幫助 LLM 判斷這個 Email 是否需要回覆

Procedural Memory: 寫在 LLM 的 System Prompt 中的內容，告訴 LLM 如何使用 Calandar Tool 以及 Email Writing Tool

Semantic Memory: 存放既定的事實 (Ex. Response Preference)
