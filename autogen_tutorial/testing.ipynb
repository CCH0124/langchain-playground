```python
local_llm_config = {
    "config_list": [
        {
            "model": "gpt-oss:20b",  # Loaded with LiteLLM command
            "api_key": "NotRequired",  # Not needed
            "base_url": "http://127.0.0.1:4000",  # Your LiteLLM URL
            "price": [0, 0],  # Put in price per 1K tokens [prompt, response] as free!
        }
    ],
    "cache_seed": None,  # Turns off caching, useful for testing different models
}
```

## Define an AutoGen agent

```python
from autogen import ConversableAgent

agent = ConversableAgent(
    name="chatbot",
    llm_config=local_llm_config,
    human_input_mode="NEVER",
)
```

```python
reply = agent.generate_reply(
    messages=[{"content": "Tell me a joke.", "role": "user"}]
)
print(reply)
```

```python
reply = agent.generate_reply(
    messages=[{"content": "Repeat the joke.", "role": "user"}]
)
print(reply)
```

## Conversation

Setting up a conversation between two agents, Cathy and Joe, where the memory of their interactions is retained.

```python
itachi = ConversableAgent(
    name="Itachi",
    system_message=
    "Your name is Itachi and you are a stand-up comedian.",
    llm_config=local_llm_config,
    human_input_mode="NEVER",
)

naruto = ConversableAgent(
    name="Naruto",
    system_message=
    "Your name is Naruto and you are a stand-up comedian. "
    "Start the next joke from the punchline of the previous joke.",
    llm_config=local_llm_config,
    human_input_mode="NEVER",
)
```

```python
chat_result = naruto.initiate_chat(
    recipient=itachi, 
    message="I'm Naruto. Itachi, let's keep the jokes rolling.",
    max_turns=2,
)
```

```python
import pprint

pprint.pprint(chat_result.chat_history)
```

```python
pprint.pprint(chat_result.cost)
```

```python
pprint.pprint(chat_result.summary)
```

## Get a better summary of the conversation

```python
chat_result = naruto.initiate_chat(
    itachi, 
    message="I'm Naruto. Itachi, let's keep the jokes rolling.", 
    max_turns=2, 
    summary_method="reflection_with_llm",
    summary_prompt="Summarize the conversation",
)
```

```python
pprint.pprint(chat_result.summary)
```

## Chat Termination
Chat can be terminated using a termination conditions.

```python
cathy = ConversableAgent(
    name="cathy",
    system_message=
    "Your name is Cathy and you are a stand-up comedian. "
    "When you're ready to end the conversation, say 'I gotta go'.",
    llm_config=local_llm_config,
    human_input_mode="NEVER",
    is_termination_msg=lambda msg: "I gotta go" in msg["content"],
)

joe = ConversableAgent(
    name="joe",
    system_message=
    "Your name is Joe and you are a stand-up comedian. "
    "When you're ready to end the conversation, say 'I gotta go'.",
    llm_config=local_llm_config,
    human_input_mode="NEVER",
    is_termination_msg=lambda msg: "I gotta go" in msg["content"] or "Goodbye" in msg["content"],
)
```

```python
chat_result = joe.initiate_chat(
    recipient=cathy,
    message="I'm Joe. Cathy, let's keep the jokes rolling."
)
```

```python
cathy.send(message="What's last joke we talked about?", recipient=joe)
```
